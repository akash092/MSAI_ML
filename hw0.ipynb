```python
# Make a scatterplot of 'mean compactness' and 'mean perimeter'. Try to color the points by
# their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot
# (from seaborn) useful for this.
import matplotlib.pyplot as plt
import seaborn as sns

# Add a column with the target labels
df_plot = df.copy()
df_plot['target'] = cancer.target_names[cancer.target]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_plot, x='mean compactness', y='mean perimeter', hue='target')
plt.title('Scatterplot of Mean Compactness vs Mean Perimeter')
plt.xlabel('Mean Compactness')
plt.ylabel('Mean Perimeter')
plt.show()
```

```python
# Make pairwise scatterplots of 'mean texture', 'mean perimeter', 'mean compactness', and 'mean
# radius'. You may find scatter_matrix (from matplotlib) or pairplot (from seaborn) useful
# for this. Do you notice that two of these features seem to be particularly correlated? Why
# might this be?
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df_plot, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='target')
plt.show()
```
